```python
from google.colab import files
import pandas as pd
```

```python
encoding_type = "utf-8-sig"
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/ideaOwl/betacityyeg-dive-opentable/main/data/2020-2021vs2019_Seated_Diner_Data.csv', encoding=encoding_type)
```

```python
df_unpivoted = data.melt(id_vars=['Type', 'Name'], var_name='date', value_name='change_from_2019')
df_unpivoted
```

```python
# Check that Montreal is spelt correctly
df_unpivoted[df_unpivoted['Name'].str.contains('Montr')]
```

```python
df_unpivoted.to_csv('unpivoted-opentable-seated-diner-data.csv', encoding=encoding_type)
```
